```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

##Dataset

```python
data=pd.read_csv("new.csv")
data.head(10)
```

##Data Cleanign

```python
missing_values = data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python

data.drop_duplicates(inplace=True)
```

```python
data.drop(['New deaths', 'New recovered'], axis=1, inplace=True)
```

```python
# Calculate fatality rate and recovery rate
data['FatalityRate'] = (data['Deaths'] / data['Confirmed']) * 100
data['RecoveryRate'] = (data['Recovered'] / data['Confirmed']) * 100
```

```python
data.head()
```

##Ploting top 10 countrys with highest confirmed cases

```python
data_sorted = data.sort_values(by='cases', ascending=False)
top_10_countries = data_sorted.head(10)
plt.figure(figsize=(10, 6))
plt.bar(top_10_countries['country'], top_10_countries['cases'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.title('All the countrys')
plt.xticks(rotation=45, ha='right')  #for roteting
plt.tight_layout()
plt.show()
```

##Ploting top 10 countrys with highest confirmed and death cases

```python
data.head(10)
```

```python
data_sorted = data.sort_values(by='cases', ascending=False)

top_10_countries = data_sorted.head(10)

plt.figure(figsize=(12, 8))
bars1 = plt.bar(top_10_countries['country'], top_10_countries['cases'], color='skyblue', label='Confirmed Cases')
bars2 = plt.bar(top_10_countries['country'], top_10_countries['death'], color='salmon', label='Deaths')
plt.xlabel('Country/Region')
plt.ylabel('Count')
plt.title('Top 10 Countries with Highest Confirmed COVID-19 Cases and Deaths')
plt.xticks(rotation=45, ha='right')  # Rotate country names for better readability
plt.legend()

# Add labels on top of bars
for bar1, bar2 in zip(bars1, bars2):
    plt.text(bar1.get_x() + bar1.get_width() / 2, bar1.get_height(), int(bar1.get_height()), ha='center', va='bottom')
    plt.text(bar2.get_x() + bar2.get_width() / 2, bar2.get_height(), int(bar2.get_height()), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

##Ploting top 10 countrys with highest confirmed with recovered

```python
data_sorted = data.sort_values(by='Confirmed', ascending=False)

top_10_countries = data_sorted.head(10)

plt.figure(figsize=(12, 8))
bars1 = plt.bar(top_10_countries['Country/Region'], top_10_countries['Confirmed'], color='skyblue', label='Confirmed Cases')
bars2 = plt.bar(top_10_countries['Country/Region'], top_10_countries['Recovered'], color='green', label='Recovered')
plt.xlabel('Country/Region')
plt.ylabel('Count')
plt.title('Top 10 Countries with Highest Confirmed COVID-19 Cases and Deaths')
plt.xticks(rotation=45, ha='right')  # Rotate country names for better readability
plt.legend()

# Add labels
for bar1, bar2 in zip(bars1, bars2):
    plt.text(bar1.get_x() + bar1.get_width() / 2, bar1.get_height(), int(bar1.get_height()), ha='center', va='bottom')
    plt.text(bar2.get_x() + bar2.get_width() / 2, bar2.get_height(), int(bar2.get_height()), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

##Ploting top 10 countrys with fatalityRate in pie chart

```python
data_sorted = data.sort_values(by='cases', ascending=False)

top_10_countries = data_sorted.head(10)

top_10_countries['death'] = (top_10_countries['death'] / top_10_countries['cases']) * 100

# Plot
plt.figure(figsize=(10, 8))
plt.pie(top_10_countries['death'], labels=top_10_countries['country'], autopct='%1.1f%%', startangle=140)
plt.title('Fatality Rate for Top 10 Countries with Highest Confirmed COVID-19 Cases')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is complete
plt.show()
```

##Ploting top 10 countrys with Recovery Rate in pie chart

```python
import pandas as pd
import matplotlib.pyplot as plt

data_sorted = data.sort_values(by='Confirmed', ascending=False)

top_10_countries = data_sorted.head(10)

# Plotting
plt.figure(figsize=(10, 8))
plt.pie(top_10_countries['Recovered / 100 Cases'], labels=top_10_countries['Country/Region'], autopct='%1.1f%%', startangle=140)
plt.title('Recovery Rate for Top 10 Countries with Highest Confirmed COVID-19 Cases')
plt.axis('equal')  # Equal aspect ratio ensures that pie is a circle
plt.show()
```
